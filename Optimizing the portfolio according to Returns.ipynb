```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as wb
```

```python
tickers = ['PG', 'MSFT', 'F', 'GE']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo', start='1995-1-1')['Adj Close']  
```

```python
(mydata/mydata.iloc[0]*100).plot(figsize = (15,6))
plt.show()
```

```python
returns = (mydata/mydata.shift(1)) -1
returns.head()
```

```python
#Devising portfolio for Maximising Returns 
import random
weights = np.array([0.25,0.25,0.25,0.25])
final = 0
weight_f = []
for i in range(1,1000):
    a = random.random()
    b = random.random()
    c = random.random()
    d = random.random()
    sum = a+b+c+d

    weights = [a/sum, b/sum, c/sum, d/sum];


    annual_returns = returns.mean() * 250
    annual_returns
    ret= np.dot(annual_returns, weights)
    if(final<ret):
        final=ret
        weight_f = weights
print(final)
print(weight_f)
```

```python
print('Best Returns got'+str(np.round(final*100))+'%')
```
